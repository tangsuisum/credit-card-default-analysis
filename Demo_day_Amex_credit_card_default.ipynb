```python
df['gender'] = df['gender'].map({'F':0, 'M':1})
```

```python
df['owns_car'].unique()
```

```python
df['owns_house'].unique()
```

```python
binary_columns = ['owns_car','owns_house']

for column in binary_columns:
    df[column] = df[column].map({'N':0, 'Y':1})
    
```

```python
df['occupation_type'].unique()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df = pd.get_dummies(data=df, columns=['occupation_type'])
```

```python
df.info()
```

```python
train_data =df.drop(['credit_card_default','customer_id','name'], axis =1)
train_labels = df['credit_card_default']
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=5)
model.fit(train_data,train_labels)
```

```python
import graphviz
columns = list(train_data.columns)
dot_data = tree.export_graphviz(model, out_file = None, feature_names=columns, class_names=['No','Yes'],filled = True, rounded = True)
```

```python
graph = graphviz.Source(dot_data)
graph
```

1. From the result of decision tree, the individuals or companies with credit score less than 649.5 high tendency be a credit card defaulter.

2. The individuals or companies with credit score between 649.5 and 699.5, used less than 70 credit limit are low risk credit card defaulter.

```python
from sklearn import tree
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_data, train_labels, test_size = 0.3, random_state = 1)
                                                    
```

```python
model = tree.DecisionTreeClassifier(max_depth= 4)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print ('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```
